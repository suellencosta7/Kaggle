```python
import pandas as pd

train_file = pd.read_csv('TitanicChallenger/train.csv')
test_file = pd.read_csv('TitanicChallenger/test.csv')

# Extrai os IDs de passageiros do conjunto de treinamento e teste
train_passId = train_file['PassengerId']
test_passId = test_file['PassengerId']

# Cria uma lista com 1 se o ID de passageiro do conjunto de treinamento estiver presente no conjunto de teste, senão 0
test_survived = [1 if num in test_passId.values else 0 for num in train_passId]

# Cria um DataFrame com as colunas 'Train' e 'Test'
data_frame = pd.DataFrame({'Train': train_passId, 'Test': test_passId, 'Survived in Test': test_survived})

data_frame.to_csv('arquivo_final.csv')

# Exibe o DataFrame
print(data_frame)
```
